```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def hypothesis(x_norm,weight,bias):
    return  (weight * x_norm) + bias
```

```python
def cost_function(y_pred,y_norm):
    cost_function_value =  np.sum((y_pred - y_norm) ** 2)/2
    return cost_function_value
```

```python
def plot_data(x,y):
    plt.plot(x,y,'ro',ms=3,color="blue")
    plt.xlabel("Multiplier")
    plt.ylabel("Answer")
```

```python
def Gradient_Desecent(weight,bias,alpha):
    updated_weight = weight + (np.sum((y_norm - y_pred)* x_norm)  )* alpha
    
    updated_bias = bias + np.sum((y_norm - y_pred)) * alpha
    return updated_weight,updated_bias
```
