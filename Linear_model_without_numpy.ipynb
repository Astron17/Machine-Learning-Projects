```python
def hypothesis( x , weight , bias ):
    return x * weight + bias
```

```python

def cost_function(x,y):
    cost_value = 0
    for i in range(len(x)):
        cost_value += (hypothesis(x[i],weight,bias)-y[i]) ** 2
    return cost_value/2
```

```python
def Gradient_desecent(weight,bias,alpha):
    weight_sum = 0
    bias_sum = 0
    for i in range(len(x)):
        
        weight_sum += (y[i] - hypothesis(x[i],weight,bias)) * x[i]  
        bias_sum += (y[i] - hypothesis(x[i],weight,bias))
    updated_weight = weight + weight_sum * alpha
    updated_bias = bias + bias_sum *alpha
    return updated_weight,updated_bias
    
```
